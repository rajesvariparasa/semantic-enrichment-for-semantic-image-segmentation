```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
path = r'E:\Thesis\plots\pc_labelled_data_impact.csv'
data = pd.read_csv(path)
```

```python
data
```

```python
# map names:
dict = {
    's2': 's2_scratch',
    's2_siam_96': 's2_siam_96_scratch',
    'single_recon': 'single_rec_ft',
    'single_segsiam_96': 'single_seg_siam_96_ft',
    'dual': 'dual_ft'
}
```

```python
# rename task name based on dict
data['task_name'] = data['task_name'].map(dict)
data
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Group by task_name and plot lines for each task_name
for task_name, group in data.groupby('task_name'):
    group = group.sort_values(by='pc_labelled_data')
    ax.plot(group['pc_labelled_data'], group['avg_val_iou'], marker='o', label=task_name)

ax.set_xlabel('Percentage of Labelled Data')
ax.set_ylabel('Average Validation IOU')
ax.set_title('Comparison of Tasks based on Percentage of Labelled Data')
ax.legend()
plt.grid(True)
```

```python
# plot same with avg_val_iou and error bar of stddev
fig, ax = plt.subplots(figsize=(10, 6))

# Group by task_name and plot lines for each task_name
for task_name, group in data.groupby('task_name'):
    group = group.sort_values(by='pc_labelled_data')
    ax.errorbar(group['pc_labelled_data'], group['avg_val_iou'], yerr=group['stddev_val_iou'], marker='o', label=task_name, capsize=5)

ax.set_xlabel('Percentage of Labelled Data')
ax.set_ylabel('Average Validation IOU')
ax.set_title('Comparison of Tasks based on Percentage of Labelled Data')
ax.legend()
```

```python
# combine above two plots into one with subplots
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(15, 5))

# Group by task_name and plot lines for each task_name

for task_name, group in data.groupby('task_name'):
    group = group.sort_values(by='pc_labelled_data')
    ax1.errorbar(group['pc_labelled_data'], group['avg_val_iou'], yerr=group['stddev_val_iou'], marker='o', label=task_name, capsize=5)
    ax2.plot(group['pc_labelled_data'], group['avg_test_iou'], marker='o', label=task_name)

ax1.set_xlabel('Percentage of Labelled Data')
ax1.set_ylabel('IOU')
ax1.set_title('Avg Fold Validation IoU')
ax1.set_xticks([10, 25, 50, 75, 100])
ax1.legend()
ax1.grid(True)

ax2.set_xlabel('Percentage of Labelled Data')
ax2.set_ylabel('IoU')
ax2.set_title('Test IoU')
ax2.set_xticks([10, 25, 50, 75, 100])
ax2.legend()
ax2.grid(True)
plt.suptitle('Comparison of Tasks based on Percentage of Labelled Data')
plt.show()
```
